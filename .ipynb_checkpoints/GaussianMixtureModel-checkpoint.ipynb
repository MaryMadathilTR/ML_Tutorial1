```python
import numpy as np
import pandas as pd

df = pd.read_csv('CC GENERAL.csv')
df.head()
```

```python
df.info()
```

```python
df.drop('CUST_ID',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
impute = SimpleImputer(strategy='mean')
df['MINIMUM_PAYMENTS'] = impute.fit_transform(df[['MINIMUM_PAYMENTS']])
```

```python
df.dropna(axis=0,inplace=True)
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
df_scaler = scale.fit_transform(df)
df_scaler = pd.DataFrame(df_scaler)
df_scaler
```

```python
df_scaler.columns = df.columns
df_scaler
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaler)
df_pca = pd.DataFrame(df_pca)
df_pca.columns = ['PC1','PC2']
df_pca
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
plt.figure(figsize=(8,6))
plt.hist(df_pca['PC1'],density=True,color='b')
mu,std = norm.fit(df['PC1'])
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)  # Probability density function
plt.plot(x, p, 'k', linewidth=2)  # Plot the PDF
plt.title(f'Fit results: Mean={mu:.2f}, Std={std:.2f}')
plt.xlabel('Data Values')
plt.ylabel('Probability Density')
plt.show()
```

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=3)
gmm.fit(df_pca)
```

```python

# Visualizing the clustering 
plt.scatter(df_pca['PC1'], df_pca['PC2'],  
           c = GaussianMixture(n_components = 3).fit_predict(df_pca), cmap =plt.cm.winter, alpha = 0.6) 
plt.show() 
```
